```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("gym_members_exercise_tracking.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.hist(figsize = (20,20))
plt.show()
```

```python
sns.pairplot(df)
```

```python
df.info()
```

```python
ax = sns.lmplot(x='Session_Duration (hours)', y='Calories_Burned', data=df, height=9, aspect=1.15)
ax.set(xlabel = "Session Duration (Hours)", ylabel = "Calories Burned")
plt.title("Duration vs. Calories Burned")
plt.savefig('images/duration_vs_cals.png', bbox_inches='tight', facecolor='white')
```

```python
sns.lmplot(x='Experience_Level', y='Calories_Burned', data=df, height=9, aspect=1.15)
ax.set(xlabel = "Experience Level", ylabel = "Calories Burned")
plt.title("Level vs. Calories Burned")
plt.savefig('images/level_vs_cals.png', bbox_inches='tight', facecolor='white')
```

```python
sns.lmplot(x='Fat_Percentage', y='Calories_Burned', data=df, height=9, aspect=1.15)
ax.set(xlabel = "Fat Percentage", ylabel = "Calories Burned")
plt.title("Fat Percentage vs. Calories Burned")
plt.savefig('images/fat_vs_cals.png', bbox_inches='tight', facecolor='white')
```
